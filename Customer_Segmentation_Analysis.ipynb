```python
df = pd.read_csv("ifood_df.csv")

print("Dataset Loaded Successfully!")
df.head()
```

```python
print(df.columns.tolist())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Income"], bins=30, kde=True, color="skyblue")

plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Number of Customers")

plt.show()
```

```python
marital_counts = {
    "Married": df["marital_Married"].sum(),
    "Single": df["marital_Single"].sum(),
    "Together": df["marital_Together"].sum(),
    "Divorced": df["marital_Divorced"].sum(),
    "Widow": df["marital_Widow"].sum()
}

plt.figure(figsize=(8,5))

sns.barplot(
    x=list(marital_counts.keys()),
    y=list(marital_counts.values()),
    palette="Set2"
)

plt.title("Marital Status Distribution")
plt.xlabel("Marital Status")
plt.ylabel("Count")

plt.show()
```

```python
education_counts = {
    "Graduation": df["education_Graduation"].sum(),
    "Master": df["education_Master"].sum(),
    "PhD": df["education_PhD"].sum(),
    "Basic": df["education_Basic"].sum(),
    "2nd Cycle": df["education_2n Cycle"].sum()
}

plt.figure(figsize=(8,5))

sns.barplot(
    x=list(education_counts.keys()),
    y=list(education_counts.values()),
    palette="viridis"
)

plt.title("Education Level")
plt.xlabel("Education")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Age"], bins=25, kde=True, color="orange")

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["MntTotal"], bins=30, kde=True, color="green")

plt.title("Total Spending")
plt.xlabel("Amount Spent")
plt.ylabel("Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Income",
    y="MntTotal",
    color="red"
)

plt.title("Income vs Total Spending")

plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(),
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
top10 = df.sort_values("Income", ascending=False)

top10.head(10)
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df[["Income","MntTotal"]]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

df["Cluster"] = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="Income",
    y="MntTotal",
    hue="Cluster",
    palette="Set1"
)

plt.title("Customer Segments")

plt.show()
```

```python
df.to_csv("customer_segmentation_cleaned.csv", index=False)
```
